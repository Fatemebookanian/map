```python
import folium

# ساخت نقشه اصلی
m = folium.Map(location=[30.29338, 57.08651], zoom_start=12)

# افزودن اولین تایل (تایل شما)
folium.TileLayer(
    tiles="https://fatemebookanian.github.io/map/tile/{z}/{x}/{y}.png",
    attr="Your Custom Tile",
    name="Custom Map"
).add_to(m)

# افزودن دومین تایل (تایل OSM)
folium.TileLayer(
    tiles="https://tile.openstreetmap.org/{z}/{x}/{y}.png",
    attr="OSM",
    name="Leaflet Map"
).add_to(m)

# افزودن کنترل لایه‌ها
folium.LayerControl().add_to(m)

# ذخیره نقشه در فایل HTML
m.save("map_with_tiles.html")
```

```python
import folium
from folium.plugins import DualMap

# ساخت نقشه DualMap برای مقایسه
dual_map = DualMap(location=[30.29338, 57.08651], zoom_start=12)

# افزودن لایه اول به نقشه سمت چپ
folium.TileLayer(
    tiles="https://fatemebookanian.github.io/map/tile/{z}/{x}/{y}.png",
    attr="Your Custom Tile",
    name="Custom Map"
).add_to(dual_map.m1)

# افزودن لایه دوم به نقشه سمت راست
folium.TileLayer(
    tiles="https://tiles.stadiamaps.com/tiles/alidade_satellite/{z}/{x}/{y}{r}.{ext}",
    attr="OSM",
    name="Leaflet Map"
).add_to(dual_map.m2)

# ذخیره نقشه در فایل HTML
dual_map.save("map_comparison.html")
```

```python
import folium

# ساخت نقشه مرکزی
m = folium.Map(location=[30.29338, 57.08651], zoom_start=12)

# افزودن لایه اول (Custom Map)
custom_tile = folium.raster_layers.TileLayer(
    tiles="https://fatemebookanian.github.io/map/tile/{z}/{x}/{y}.png",
    attr="Custom Map",
    name="Custom Map"
).add_to(m)

# افزودن لایه دوم (OSM)
osm_tile = folium.raster_layers.TileLayer(
    tiles="https://tile.openstreetmap.org/{z}/{x}/{y}.png",
    attr="OpenStreetMap",
    name="OSM"
).add_to(m)

# اضافه کردن پلاگین های جاوااسکریپت و CSS به فایل HTML
swipe_js = """
    <link rel="stylesheet" href="https://unpkg.com/leaflet-side-by-side@2.1.1/leaflet-side-by-side.min.css" />
    <script src="https://unpkg.com/leaflet-side-by-side@2.1.1/leaflet-side-by-side.min.js"></script>
    <script>
        document.addEventListener('DOMContentLoaded', function() {
            var leftLayer = document.querySelector('div.leaflet-pane .leaflet-tile-pane').children[0]; 
            var rightLayer = document.querySelector('div.leaflet-pane .leaflet-tile-pane').children[1];
            var sideBySide = L.control.sideBySide(leftLayer, rightLayer);
            sideBySide.addTo(window.map);
        });
    </script>
"""

# اضافه کردن کد جاوااسکریپت به فایل HTML
m.get_root().html.add_child(folium.Element(swipe_js))

# ذخیره نقشه در فایل HTML
m.save("map_swiper.html")
```
